```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("bangladesh-data.xlsx")
data = data["Total DASS score"]
categories_count = [0 for _ in range(13)]
for i in range(13):
    is_first_count = 1 if i else 0 
    value_count = data[(data>(i*10+is_first_count)) & (data<(i+1)*10)].count()
    categories_count[i] = value_count
    
categories_count
fig, ax = plt.subplots()
plt.bar(np.arange(13), categories_count)
for i, value in enumerate(categories_count):
    shift = 0.4 if value>99 else 0.3
    plt.text(i-shift, value+3, s=str(value))
labels = [str(i*10+1) +"-"+str((i+1)*10) for i in range(13)]
labels[0] = "0-10"
ax.set_xticks(np.arange(13), labels, rotation=45)
ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=.2)
ax.set_ylim((0, max(categories_count)+25) )
plt.ylabel("Liczba studentów")
plt.xlabel("Wynik w DASS")
plt.title("Rozkład DASS wśród studentów Bangladeszu")
plt.savefig("Rozkład_DASS.png", bbox_inches = "tight")

plt.show()
```
