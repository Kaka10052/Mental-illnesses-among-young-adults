```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("prevalence-of-alcohol-use-disorders-by-age.csv")
data =data[data["Code"] == "POL"]
column_names = ['Enitity', "Code", "Year", "10-14", "15-19 lat", "Wszyscy", "25-29 lat", "30-34 lat", "20-24 lat", "70+", "15-49", "5-14", "50-69", "Std"]
data = data.set_axis(column_names, axis=1)
data = data.reset_index()
data = data[["15-19 lat", "20-24 lat", "25-29 lat", "30-34 lat", "Wszyscy"]]
data.head()
```

```python
fig, ax = plt.subplots()
N = len(data.index)
x = np.arange(N)
for column in data.keys():
    y = data[column].tolist()
    plt.plot(x, y, 'o-', label=column)    
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.08))
xlabels = [i+1990 for i in range(0, N, 5)]
ax.set_xticks(np.arange(0, N, 5), xlabels)
ax.set_xlabel("Lata")
yticks = np.arange(1, 3.6, 0.5)
ylabels = [str(i)+"%" for i in yticks]
ax.set_yticks(yticks, ylabels)
ax.set_ylabel("Udział procentowy w całej populacji")
plt.title("Występowanie alkoholizmu względem wieku \nna przestrzeni lat 1990-2017")
plt.savefig("Występowanie_alkoholizmu.png")
plt.show()
```
